## Matrix Factorization

```python
matrix_=df.pivot_table(index='Products.id',columns='Products.name',values='Orders.total').fillna(0)
matrix_.shape
```

```python
matrix_.head()
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
x = matrix_.values.T
```

```python
SVD = TruncatedSVD(n_components=12, random_state=42)
matrix = SVD.fit_transform(x)
matrix.shape
```

```python
corr = np.corrcoef(matrix)
corr
```

```python
title= matrix_.columns
title_list=list(title)
samia = title_list.index('Emesis Bags,Blue,36.000 OZ')
corr_samia  = corr[samia]
list(title[(corr_samia >= 0.9)])
```

## Cosine Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
sparse_matrix=vect.fit_transform(df)
```

```python
doc_term_matrix = sparse_matrix.todense()
df = pd.DataFrame(doc_term_matrix, 
                  columns=vect.get_feature_names())
df
```

```python
df.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
pd.DataFrame(cosine_similarity(df,df))
```
